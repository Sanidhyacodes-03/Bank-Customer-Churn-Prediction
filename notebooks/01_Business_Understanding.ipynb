# Business Problem

Customer churn is one of the most significant challenges faced by retail banks. Acquiring a new customer is substantially more expensive than retaining an existing one. However, many banks continue to rely on reactive retention strategies, intervening only after customers exhibit clear signs of leaving.

This reactive approach often leads to increased customer attrition, higher marketing costs, and reduced customer lifetime value. To address this challenge, the bank requires a predictive analytics solution capable of identifying customers who are at high risk of churn before they actually leave.

By leveraging historical customer data and machine learning techniques, the bank can proactively engage at-risk customers through personalized retention strategies, thereby improving customer satisfaction and long-term profitability.

# Dataset Description

The dataset contains customer demographic, financial, and behavioral information collected by a European retail bank.

### Dataset Features

| Feature | Description |
|---------|-------------|
| CustomerId | Unique customer identifier |
| Surname | Customer surname |
| CreditScore | Customer credit score |
| Geography | Customer country (France, Germany, Spain) |
| Gender | Male/Female |
| Age | Customer age |
| Tenure | Years with the bank |
| Balance | Account balance |
| NumOfProducts | Number of bank products |
| HasCrCard | Credit card ownership |
| IsActiveMember | Active membership status |
| EstimatedSalary | Annual estimated salary |
| Exited | Target variable (1 = Churn, 0 = Retained) |
